```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

data['monthly_charges'] = pd.to_numeric(data['monthly_charges'], errors='coerce')
data['tenure'] = pd.to_numeric(data['tenure'], errors='coerce')
data['total_charges'] = pd.to_numeric(data['total_charges'], errors='coerce')

data_clean_clusters = data[
    (data['monthly_charges'] <= 150) & 
    (data['tenure'] <= 72)
].copy()

data_clean_clusters = data_clean_clusters.dropna(subset=['monthly_charges', 'tenure', 'total_charges'])

df_model = data_clean_clusters.copy()

df_model['signup_year'] = df_model['signup_date'].dt.year
df_model['signup_month'] = df_model['signup_date'].dt.month
df_model = df_model.drop(columns=['signup_date'])

df_model['churn'] = df_model['churn'].astype(str).str.lower().str.strip()
y = df_model['churn'].map({'yes': 1, 'no': 0})

X = df_model.drop(columns=['customer_id', 'churn'])

X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

scaler = StandardScaler()
num_cols = ['tenure', 'monthly_charges', 'total_charges', 'signup_year', 'signup_month']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

rf_clf = RandomForestClassifier(random_state=42, n_estimators=100)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title(f'Logistic Regression CM\nAccuracy: {accuracy_score(y_test, y_pred_lr):.2%}')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')
axes[0].set_xticklabels(['No Churn', 'Churn'])
axes[0].set_yticklabels(['No Churn', 'Churn'])

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=False)
axes[1].set_title(f'Random Forest CM\nAccuracy: {accuracy_score(y_test, y_pred_rf):.2%}')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')
axes[1].set_xticklabels(['No Churn', 'Churn'])
axes[1].set_yticklabels(['No Churn', 'Churn'])

plt.tight_layout()
plt.show()

print("=== LOGISTIC REGRESSION REPORT ===")
print(classification_report(y_test, y_pred_lr, target_names=['No Churn', 'Churn']))

print("\n=== RANDOM FOREST REPORT ===")
print(classification_report(y_test, y_pred_rf, target_names=['No Churn', 'Churn']))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# --- 1. CONVERT TO NUMERIC & FILTER OUTLIERS ---
data['monthly_charges'] = pd.to_numeric(data['monthly_charges'], errors='coerce')
data['tenure'] = pd.to_numeric(data['tenure'], errors='coerce')
data['total_charges'] = pd.to_numeric(data['total_charges'], errors='coerce')

data_clean_clusters = data[
    (data['monthly_charges'] <= 150) & 
    (data['tenure'] <= 72)
].copy()

data_clean_clusters = data_clean_clusters.dropna(subset=['monthly_charges', 'tenure', 'total_charges'])
df_model = data_clean_clusters.copy()

# --- 2. DATA PREPARATION & FEATURE ENGINEERING ---
df_model['signup_year'] = df_model['signup_date'].dt.year
df_model['signup_month'] = df_model['signup_date'].dt.month
df_model = df_model.drop(columns=['signup_date'])

df_model['churn'] = df_model['churn'].astype(str).str.lower().str.strip()
y = df_model['churn'].map({'yes': 1, 'no': 0})

X = df_model.drop(columns=['customer_id', 'churn'])
X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

scaler = StandardScaler()
num_cols = ['tenure', 'monthly_charges', 'total_charges', 'signup_year', 'signup_month']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])


# --- 3. UPDATED MODEL TRAINING ---
# Logistic Regression with balanced weights
log_reg = LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

# SWITCHED TO 'balanced_subsample' to force Random Forest trees to balance out clusters
rf_clf = RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced_subsample', max_depth=10)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)


# --- 4. EVALUATION VIA CONFUSION MATRICES ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title(f'Balanced Logistic Regression\nAccuracy: {accuracy_score(y_test, y_pred_lr):.2%}')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')
axes[0].set_xticklabels(['No Churn', 'Churn'])
axes[0].set_yticklabels(['No Churn', 'Churn'])

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=False)
axes[1].set_title(f'Optimized Random Forest\nAccuracy: {accuracy_score(y_test, y_pred_rf):.2%}')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')
axes[1].set_xticklabels(['No Churn', 'Churn'])
axes[1].set_yticklabels(['No Churn', 'Churn'])

plt.tight_layout()
plt.show()

print("=== BALANCED LOGISTIC REGRESSION REPORT ===")
print(classification_report(y_test, y_pred_lr, target_names=['No Churn', 'Churn']))

print("\n=== OPTIMIZED RANDOM FOREST REPORT ===")
print(classification_report(y_test, y_pred_rf, target_names=['No Churn', 'Churn']))
```

```python
!pip install lightgbm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import lightgbm as lgb

data['monthly_charges'] = pd.to_numeric(data['monthly_charges'], errors='coerce')
data['tenure'] = pd.to_numeric(data['tenure'], errors='coerce')
data['total_charges'] = pd.to_numeric(data['total_charges'], errors='coerce')

data_clean_clusters = data[
    (data['monthly_charges'] <= 150) & 
    (data['tenure'] <= 72)
].copy()

data_clean_clusters = data_clean_clusters.dropna(subset=['monthly_charges', 'tenure', 'total_charges'])
df_model = data_clean_clusters.copy()

df_model['signup_year'] = df_model['signup_date'].dt.year
df_model['signup_month'] = df_model['signup_date'].dt.month
df_model = df_model.drop(columns=['signup_date'])

df_model['churn'] = df_model['churn'].astype(str).str.lower().str.strip()
y = df_model['churn'].map({'yes': 1, 'no': 0})

X = df_model.drop(columns=['customer_id', 'churn'])
X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

scaler = StandardScaler()
num_cols = ['tenure', 'monthly_charges', 'total_charges', 'signup_year', 'signup_month']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

scale_weight = (len(y_train) - sum(y_train)) / sum(y_train)

lgb_model = lgb.LGBMClassifier(
    random_state=42,
    n_estimators=150,
    learning_rate=0.05,
    max_depth=6,
    scale_pos_weight=scale_weight,  
    verbosity=-1
)

lgb_model.fit(X_train, y_train)
y_pred_lgb = lgb_model.predict(X_test)

plt.figure(figsize=(6, 5))
cm_lgb = confusion_matrix(y_test, y_pred_lgb)
sns.heatmap(cm_lgb, annot=True, fmt='d', cmap='Oranges', cbar=False)
plt.title(f'LightGBM Confusion Matrix\nAccuracy: {accuracy_score(y_test, y_pred_lgb):.2%}')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.gca().set_xticklabels(['No Churn', 'Churn'])
plt.gca().set_yticklabels(['No Churn', 'Churn'])
plt.show()

print("=== LIGHTGBM CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred_lgb, target_names=['No Churn', 'Churn']))
```

Problem: We wanted to predict which customers are likely to churn (leave).

Result: Our LightGBM model is the most effective, catching 22% of all churners.

Recommendation: Marketing should focus retention efforts on customers with "Month-to-month" contracts and high monthly charges, as these are your highest-risk groups.
